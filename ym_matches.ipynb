### **The generational matches of our alkaline degradation of glucose computed set (5 gen) to Yang & Montgomery 1996. **

```python
import mols2grid
import pandas as pd
url = 'https://raw.githubusercontent.com/ssiddhantsharma/match-viz/main/young_and_mont_matched.csv'
df = pd.read_csv(url, sep=',', header=0)
Generation = df['Generation'].tolist()
SMILES = df['SMILES'].tolist()
LogP= df['cLogP'].tolist()
PSA= df['Polar Surface Area'].tolist()
Mass= df['Monoisotopic Mass'].tolist()
df = pd.DataFrame({"SMILES": SMILES, "id": Generation,"Molar Mass": Mass, "cLogP": LogP, "PSA": PSA})
mg = mols2grid.MolGrid(df, smiles_col="SMILES", size=(210, 130), useSVG=True, coordGen=True, fixedBondLength=40)
mg.display(subset=["id", "img"], n_cols=5, tooltip=["SMILES", "Molar Mass", "cLogP", "PSA"], style={"cLogP": lambda x: "color: red"}, tooltip_placement="bottom", fontsize="14pt", hover_color="#e3c7cb", n_rows=4, selection=False)

```
