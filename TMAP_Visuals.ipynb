```python
from faerun import Faerun
from rdkit.Chem import AllChem
import pandas as pd
url = 'https://raw.githubusercontent.com/ssiddhantsharma/match-viz/main/FORM.csv' #importing the descriptors
df = pd.read_csv(url, sep=',', header=0)
df.head(100)
```

```python
SMILES = df['SMILES'].tolist()
cLogP= df['cLogP'].tolist()
M_Complexity= df['Molecular Complexity'].tolist()
M_Flexibility= df['Molecular Flexibility'].tolist()
```

```python
!pip install mhfp
from mhfp.encoder import MHFPEncoder
import tmap
# The number of permutations used by the MinHashing algorithm
perm = 512

# Initializing the MHFP encoder with 512 permutations
enc = MHFPEncoder(perm)

# Create MHFP fingerprints from SMILES
# The fingerprint vectors have to be of the tm.VectorUint data type
fingerprints = [tmap.VectorUint(enc.encode(s)) for s in df["SMILES"]]
```

```python
lf = tmap.LSHForest(perm)

# Add the Fingerprints to the LSH Forest and index
lf.batch_add(fingerprints)
lf.index()
# Get the coordinates
x, y, s, t, _ = tmap.layout_from_lsh_forest(lf)
```

```python
# Now plot the data
faerun = Faerun(view="front", coords=False)
faerun.add_scatter(
    "Descriptors",
    {   "x": x, 
        "y": y, 
        "c": [cLogP, M_Complexity, M_Flexibility], 
        "labels": df["SMILES"]},
    point_scale=2.5, shader="smoothCircle", max_point_size=20,
    colormap = ["Spectral", "Spectral", "viridis"],
    has_legend=True, 
    categorical=[False, False, False,],
    series_title = ['cLogP', 'M_Complexity', 'M_Flexibility'],
    legend_labels = [None, None, None],
)
faerun.add_tree("Descriptors_tree", {"from": s, "to": t}, point_helper="Descriptors", color="#666666")
# Choose the "smiles" template to display structure on hover
faerun.plot('Formose_Reaction_Descriptors', template="smiles", notebook_height=750)
```
